## Inner join

```python
# default the python takes inner join
'''
party and product are the two columns that are common in the two join so it took it and all the values which common in both the table
matched items are printed
'''
```

```python
filter_2017.merge(filter_2018)
```

```python
#when the basis is changed only on the party it divided the product
```

```python
filter_2017.merge(filter_2018,on="Party")
```

## outer join

```python
'''
outer join combines whole dataframe and combines into one  on the basis of part and product so when there no similar values
it gives NaN'''
```

```python
filter_2017.merge(filter_2018,how="outer")
```

```python
# left  join
```

```python
filter_2017.merge(filter_2018,how="left")
```

```python
# right join
```

```python
filter_2017.merge(filter_2018,how="right")
```

# groupby

```python
df=pd.read_csv("Flavors.csv")
df
```

```python
df_groupby_frame = df.groupby("Base Flavor")
```

```python
df_groupby_frame.mean(numeric_only = True)
```

```python
df_groupby_frame.count()
```

```python
df_groupby_frame.min()
```

```python
df_groupby_frame.max()
```

```python
df.groupby("Base Flavor").agg({"Flavor Rating":["mean","min","max","count","std"],"Texture Rating":["mean","min","max","count","std"]})
```
